```python
import pandas as pd
df = pd.read_csv('data/2023nba.csv')
df
```

```python
# NaN 제거
df = df.dropna(how = 'any')
df
```

```python
pos_counts = df['POS'].value_counts()
print(pos_counts)
```

```python
df.describe().T
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(df['Min'].dropna(), bins=20, edgecolor='black')
plt.title('Min Data Distribution')
plt.xlabel('Min')
plt.ylabel('Frequency')
plt.show()
```

```python
columns_to_drop = ['PName','Team','GP','W','L','FP','DD2','TD3','+/-']
df = df.drop(columns=columns_to_drop)
df
```

```python
df_filtered = df[df['Min'] >= 100]
df_filtered
```

```python
columns_to_drop = ['PTS','FGM','3PM','FTM','FTA','FT%']
df_filtered = df_filtered.drop(columns=columns_to_drop)
df_filtered
```

```python
columns_to_divide = ['FGA', '3PA', 'OREB', 'DREB', 'REB', 'AST', 'TOV', 'STL', 'BLK', 'PF']

for column in columns_to_divide:
    df_filtered[column] = df_filtered[column] / df_filtered['Min']
    
df_filtered
```

```python
columns_to_drop = ['Age','Min']
df_filtered = df_filtered.drop(columns=columns_to_drop)
df_filtered
```

```python
columns_to_drop = ['FG%','3P%','REB']
df_filtered = df_filtered.drop(columns=columns_to_drop)
df_filtered
```

```python
columns_to_drop = ['FGA','TOV','STL','PF']
df_filtered = df_filtered.drop(columns=columns_to_drop)
df_filtered
```

```python
import pandas as pd

df_filtered = df_filtered[~df_filtered['POS'].isin(['SF', 'SG', 'PF', 'PG'])]

df_filtered = df_filtered.reset_index(drop=True)
df_filtered
```

```python
import pandas as pd

pos_labels = {
    'C': 0,
    'F': 1,
    'G': 2
}

df_filtered['POS'] = df_filtered['POS'].map(pos_labels)

df_filtered
```

```python
pos_counts = df_filtered['POS'].value_counts()
print(pos_counts)
```

```python
pos_averages = df_filtered.groupby('POS').mean()

print(pos_averages*100)
```

```python
from sklearn.preprocessing import StandardScaler

df_scaled = StandardScaler().fit_transform(df_filtered.iloc[:,1:]) # 첫번째 열 인덱스 0에 해당 -> POS 제외
print(df_scaled)
```

```python
df_filtered.POS
```

```python
#LDA 진행

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2) 
lda.fit(df_scaled, df_filtered.POS)
df_lda = lda.transform(df_scaled)

df_nba = pd.DataFrame(df_lda, columns=['comp1', 'comp2'])
df_nba['target'] = df_filtered.POS

df_nba
```

```python
markers = ['^','s','o','*','v']

class_labels = ['C', 'F', 'G','4','5']
plt.figure(figsize=(10, 7))

for i, marker in enumerate(markers):
  x = df_nba[df_nba['target'] == i].iloc[:, 0]
  y = df_nba[df_nba['target'] == i].iloc[:, 1]
  plt.scatter(x, y, marker=marker, label=class_labels[i])
plt.legend()
plt.show()
```

```python
#K - mean

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, n_init=10, random_state=0)
kmeans.fit(df_lda)

lda_kmeans_df = pd.DataFrame(df_lda, columns=['comp1', 'comp2'])
lda_kmeans_df['cluster'] = kmeans.labels_

lda_kmeans_df.head()
```

```python
plt.figure(figsize=(10, 7))
plt.title("K-Means", fontsize=20)

for i in range(5): 
    x = lda_kmeans_df[lda_kmeans_df['cluster'] == i]['comp1']
    y = lda_kmeans_df[lda_kmeans_df['cluster'] == i]['comp2']
    plt.scatter(x, y, marker=markers[i], label=f'Cluster {i}')

plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
cluster_labels = kmeans.labels_

score = silhouette_score(df_lda, cluster_labels)

print('실루엣 점수:', score)
```

```python
from sklearn.metrics import silhouette_samples
silhouette_vals = silhouette_samples(df_lda, cluster_labels)

silhouette_df = pd.DataFrame({'label': cluster_labels, 'silhouette': silhouette_vals})

mean_silhouette_vals_by_cluster = silhouette_df.groupby('label').mean()

print(mean_silhouette_vals_by_cluster)
```

```python
# Mean - Shift use estimate_bandwidth

from sklearn.cluster import MeanShift, estimate_bandwidth
import numpy as np

bandwidth = estimate_bandwidth(df_lda, quantile=0.2, n_samples=200)
print('최적 bandwidth값: ',round(bandwidth,3))

meanshift = MeanShift(bandwidth=bandwidth, bin_seeding=True)
meanshift.fit(df_lda)

# 클러스터의 수 확인
n_clusters_ = len(np.unique(meanshift.labels_))
print(f'number of clusters: {n_clusters_}')

lda_meanshift_df = pd.DataFrame(df_lda, columns=['comp1', 'comp2'])
lda_meanshift_df['cluster'] = meanshift.labels_

plt.figure(figsize=(10, 7))
plt.title(f"Mean Shift", fontsize=20)

for i in range(n_clusters_): 
    # 왜 range(2)가 아니고 n_clusters 인가요? -> estimate_bandwitdh는 클러스터의 수를 지정하지 않기 때문
    x = lda_meanshift_df[lda_meanshift_df['cluster'] == i]['comp1']
    y = lda_meanshift_df[lda_meanshift_df['cluster'] == i]['comp2']
    plt.scatter(x, y, marker=markers[i],label=f'Cluster {i}')

plt.legend()
plt.show()
```

```python
#GMM

from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3, random_state=0)

gmm.fit(df_lda) # 주어진 데이터에 대해 각 가우시안 분포의 매개변수와 각 클러스터에 속할 확률을 학습

gmm_labels = gmm.predict(df_lda) # 학습된 모델을 바탕으로, 각 데이터 포인트가 속할 확률이 가장 높은 클러스트를 예측

gmm_df = pd.DataFrame(df_lda, columns=['comp1', 'comp2'])
gmm_df['cluster'] = gmm_labels

plt.figure(figsize=(10, 7))
plt.title("GMM", fontsize=20)

for i in range(3):
    x = gmm_df[gmm_df['cluster'] == i]['comp1']
    y = gmm_df[gmm_df['cluster'] == i]['comp2']
    plt.scatter(x,y,marker=markers[i],label=f'Cluster {i}')

plt.legend()
plt.show()
```

```python
#DBSCAN

from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.3, min_samples=5)

dbscan.fit(df_lda)

# 클러스터 수를 계산. set(dbscan.labels)로 전체 레이블 검색. (1 if -1 in dbscan.labels_ else 0)로 레이블 중 -1 감지시 전체 수에서 1을 뺌(Noise)
n_clusters_ = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0) 

print(f'number of clusters: {n_clusters_}')

lda_dbscan_df = pd.DataFrame(df_lda, columns=['comp1', 'comp2'])
lda_dbscan_df['cluster'] = dbscan.labels_

# 노이즈 포인트는 -1로 라벨링됨
unique_labels = set(dbscan.labels_)

plt.figure(figsize=(10, 7))
plt.title("DBSCAN", fontsize=20)

for k in unique_labels:
    class_member_mask = (dbscan.labels_ == k)
    xy = df_lda[class_member_mask]

    if k == -1:# 노이즈는 검은색으로 표시
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor='black', markeredgecolor='k', markersize=4, label='Noise')
    else:
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=None, markersize=6, label=f'Cluster {k}')

plt.legend()
plt.show()
```
